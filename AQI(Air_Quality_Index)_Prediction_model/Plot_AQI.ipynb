### In this Notebook, we are esentially going to find the average of the TV(Target variable) PM2.5 across all the 30/31 days of the Month of that specific year  

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def avg_data_2013():
    temp_i=0
    average=[]
    for rows in pd.read_csv("./Data/AQI/aqi2013.csv",chunksize=24): #Get all 24 rows at once with chunksize, which represents 24 readings in a day!!
        add_var=0
        avg=0.0
        data=[]
        #print("row = \n",rows)
        df = pd.DataFrame(data=rows)
        for index,row in df.iterrows():#iterrows() is basically extracts each of the row data our of the extracted 24 rows above
            data.append(row['PM2.5'])
        for i in data:
            if type(i) is float or type(i) is int:
                add_var+=i
            elif type(i) is str:
                if i!='NoData' and i!='PwrFail' and i!='---' and i!='INVld':
                    temp=float(i)
                    add_var=add_var+temp
        avg = add_var/24
        temp_i=temp_i+1
        
        average.append(avg)
        return average
    
```

```python

def get_avg(yr):
    temp_i=0
    average=[]
    csv="./Data/AQI/aqi{0}.csv".format(yr)
    for rows in pd.read_csv(csv,chunksize=24): #Get all 24 rows at once with chunksize, which represents 24 readings in a day!!
        add_var=0
        avg=0.0
        data=[]
        #print("row = \n",rows)
        df = pd.DataFrame(data=rows)
        for index,row in df.iterrows():#iterrows() is basically extracts each of the row data our of the extracted 24 rows above
            data.append(row['PM2.5'])
        for i in data:
            if type(i) is float or type(i) is int:
                add_var+=i
            elif type(i) is str:
                if i!='NoData' and i!='PwrFail' and i!='---' and i!='InVld':
                    temp=float(i)
                    add_var=add_var+temp
        avg = add_var/24
        temp_i=temp_i+1
        
        average.append(avg)
    return average
```

```python
lst_yr=[2013,2014,2015,2016,2017,2018]
yr=[]
for i in lst_yr:
    #yr="lst_{0}".format(i)
    yr.append(get_avg(i))
```
